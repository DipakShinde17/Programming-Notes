## Broadcasting

```python
arr1 = np.array([[10,20,30,40],
                 [50,60,70,80]])
arr2 = np.array([1,2,3,4])
```

```python
arr3 = arr1 + arr2
arr3
```

```python
arr1 = np.array([1,2,3,4,5,6,7])
arr2 = np.array([1,2,3])
arr3 = arr1+arr2
arr3
```

```python
arr2 = np.pad(arr2,(0,4))
```

```python
arr2
```

```python
arr3 = arr1 + arr2
```

```python
arr3
```
