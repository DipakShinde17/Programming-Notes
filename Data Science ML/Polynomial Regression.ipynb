```python
import pandas as pd
import numpy as np 
```

```python
data = pd.read_csv("Position_Salaries.csv")
```

```python
data.head()
```

```python
X = data[["Level"]]
y = data["Salary"]
```

```python
### check with linear regression 
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X,y)
```

```python
pred = lr.predict(X)
```

```python
from sklearn.metrics import r2_score
r2_score(y,pred)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(7,3))
plt.scatter(data["Level"],data["Salary"])
plt.plot(data["Level"],pred)
```

### Implementing polynomial regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=4)
```

```python
x_trans = poly.fit_transform(X)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_trans,y)
```

```python
poly_predict = lr.predict(x_trans)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_trans,y)
```

```python
poly_predict = lr.predict(x_trans)
```

```python
r2_score(y,poly_predict)
```

```python
plt.figure(figsize=(7,3))
plt.scatter(data["Level"],data["Salary"])
plt.plot(data["Level"],poly_predict)
```
