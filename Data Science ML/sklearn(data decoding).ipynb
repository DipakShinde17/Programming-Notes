```python
from sklearn.preprocessing import OneHotEncoder
```

```python
data = pd.DataFrame({"Type":["Butterscpoth","Vanilla","Chocolate","Kaju","Butterscpoth","Kaju","Butterscpoth","Vanilla","Kaju","Vanilla"],
                     "Size":["Small","Medium","Large","Small","Large","Medium","Small","Large","Medium","Medium"],
                     "Qnt":[2,3,4,5,1,2,3,5,3,2]
    
})
```

```python
data.head()
```

```python
ohe = OneHotEncoder()
result = ohe.fit_transform(data[["Type"]]).toarray()
```

```python
result
```

```python
result = pd.DataFrame(result,columns = ["Butterscpoth","Vanilla","Chocolate","Kaju"])
result
```
