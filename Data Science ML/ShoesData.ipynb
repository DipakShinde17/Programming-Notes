```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(x = "sold_qty",data = data,kde = True)
mean = data["sold_qty"].mean()
std = data["sold_qty"].std()
plt.axvline(mean,color = "red")
plt.axvline(mean-3*std,color = "green")
plt.axvline(mean+3*std,color = "green")

std3 = mean+3*std
std3
```

```python
plt.boxplot(data["sold_qty"])
plt.show
```

```python
Q1 = data["sold_qty"].quantile(0.25)
Q3 = data["sold_qty"].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data["sold_qty"])
plt.title("Outliers using IQR")
plt.show()
```
