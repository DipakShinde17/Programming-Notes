```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
text = '''Dr. Smith lives in the U.S. and works at New York University. 
He specializes in A.I. and has published over 3.14 papers every year. 
He often travels to Washington D.C. for conferences, workshops, etc. 
Recently, he received an award from the A.I. Society.'''
```

```python
doc = nlp(text)
```

```python
type(doc)
```

```python
for s in doc.sents:
    print(s)
```

```python
text = "Hello! Are you coming?"
```

```python
doc = nlp(text)
```

```python
for token in doc:
    print(token," - ",token.pos_)
```

```python
text = "My mailid is sonal@gmail.com"
doc = nlp(text)
```

```python
for token in doc:
    if(token.like_email):
        print(token)
```

# Stemming/Lemmatization

```python
import nltk
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
words = ["eating", "eats", "ate", "adjustable", "rafting", "ability", "meeting","better"]
```

```python
for word in words:
    print(word,"-",ps.stem(word))
```

```python
text = "I am meeting him tomorrow at the meeting"
words = text.split(" ")
words
```

```python
for word in words:
    print(word,"   ",ps.stem(word))
```

```python
text
```

```python
doc = nlp(text)
for token in doc:
    print(token,"   ", token.lemma_,"   ",token.pos_)
```

```python
words
```

```python
text = " ".join(words)
text
```

```python
doc = nlp(text)
for token in doc:
    print(token,"   ", token.lemma_)
```

```python
text = "Tesla Inc has acquired Twitter Inc in India"
text = "Mr. Smith is a wise person"
doc = nlp(text)
for token in doc.ents:
    print(token,"  ",token.label_)
```
