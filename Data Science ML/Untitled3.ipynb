```python
import pandas as pd 
```

```python
data = pd.DataFrame({
     "EXP":[1,2,3,4,5,6,7,8,9],
    "SALARY":[1000,2000,4500,8000,9000,1200,3000,1500,2000]})
```

```python
data
```

```python
X = data[["EXP"]]
y = data["SALARY"]
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X,y)
```

```python
pred = lr.predict(X)
```

```python
pred
```

```python
from sklearn.metrics import r2_score
r2_score(y,pred)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(7,3))
plt.scatter(data["EXP"],data["SALARY"])
plt.plot(data["EXP"],pred)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=8)
```

```python
x_trans = poly.fit_transform(X)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_trans,y)
```

```python
poly_predict = lr.predict(x_trans)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_trans,y)
```

```python
poly_predict = lr.predict(x_trans)
```

```python
r2_score(y,poly_predict)
```

```python
plt.figure(figsize=(7,3))
plt.scatter(data["EXP"],data["SALARY"])
plt.plot(data["EXP"],poly_predict)
```
