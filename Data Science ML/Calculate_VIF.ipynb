## VIF

```python
import pandas as pd
```

```python
data = pd.read_csv("salaries for VIF.csv")
```

```python
data = data.drop("Salary",axis = 1)
```

```python
data.head()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def calculateVif(data):
    vif = []
    output = pd.DataFrame()
    output["Attributes"] = data.columns
    for i in range(0,data.shape[1]):
        vif.append(variance_inflation_factor(data,i))
    output["vif"] = vif
    return output
```

```python
calculateVif(data)
```

```python
data = data.drop("Age",axis = 1)
```

```python
calculateVif(data)
```
