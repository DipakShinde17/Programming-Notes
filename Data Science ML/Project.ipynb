### Fill Outlier

```python
Q1 = data["skills_match_score"].quantile(0.25)
Q3 = data["skills_match_score"].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

lower_limit, upper_limit
```

```python
data["skills_match_score"] = data["skills_match_score"].clip(lower=lower_limit, upper=upper_limit)
```

```python
def showPlot():    
    plt.subplot(1, 2, 1)
    plt.boxplot(data["skills_match_score"])
    plt.subplot(1, 2, 2)
    sns.histplot(data = data, x = data["skills_match_score"], kde=True)
    plt.axvline(data["skills_match_score"].mean()+3*data["skills_match_score"].std(), color='red')
    plt.axvline(data["skills_match_score"].mean()-3*data["skills_match_score"].std(), color='red')
showPlot()
```

### Build model

# KNN

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=7)
```

```python
knn.fit(x_train, y_train)
```

```python
test_pred = knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
test_acc = accuracy_score(y_test, test_pred)
```

```python
print("Testing Accuracy : ",test_acc)
```

```python
train_pred = knn.predict(x_train)
train_acc = accuracy_score(y_train, train_pred)
print("Training Accuracy:", train_acc)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, test_pred)
```

```python
test_acc = []
train_acc = []
k = []

for i in range(3, 16, 2):
    k.append(i)
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    train_pred = knn.predict(x_train)
    test_pred = knn.predict(x_test)
    test_acc.append(accuracy_score(y_test, test_pred))
    train_acc.append(accuracy_score(y_train, train_pred))
```

```python
print(train_acc)
```

```python
print(test_acc)
```

```python
plt.plot(k, train_acc)
plt.plot(k, test_acc)
plt.legend(["Training", "Testing"])
plt.show()
```
